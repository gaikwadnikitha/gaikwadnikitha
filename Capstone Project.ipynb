```python
#Creating running averages using 'Cost' column in the project dataframe

running = 0
count = 0

# Calculating the average cost
average_cost = round(project_df['Cost'].mean(), 2)

# Replacing missing values with average cost using a for loop
for i in range(len(project_df)):
    if pd.isnull(project_df.loc[i, 'Cost']):       #isnull() function is used to check null values
        project_df.loc[i, 'Cost'] = average_cost

    #else function is used to iterate the loop if there is no missing values
    else:
        running += project_df.loc[i, 'Cost']
        count += 1

print("Updated Project DataFrame:")
print(project_df)
```

Task 5 :
Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project 
cost only to employees who have finished the projects.

```python
# Creating a copy of final dataframe for referance
final_copy = final

print("Copy of Final DataFrame: ")
final_copy
```

```python
#Using 'for' loop to creating a list of Bonus cost
#Later, inserted the list into the final dataframe
results_list=[]

for i, r in final_copy.iterrows():
    if(final_copy['Status'].at[i] == 'Finished'):             #checking each index for "Finished" 
        result = round((r['Cost']) + (r['Cost']) * 0.05, 2)    #if found, then bonus = cost + 5% of cost,also rounding the bonus upto 2 decimal points
        results_list.append(result)                            #append() will add the result in the end of the list
    else:
        result = round((r['Cost']) + (r['Cost']) * 0, 2)       #no bonus if not finished
        results_list.append(result)

final_copy.insert(9, "Bonus", results_list)                 #inserting the bonus-5% column into the final Dataframe at index 9 

print("Updated Final DataFrame: ")
final_copy
```

Task 6 : 
Demote the designation level by 1, whose projects have status “fail”. Delete the 
employees record whose designation level is above 4.

```python

new_designation = []

for i, r in final_copy.iterrows():
    #checking if the status of the project is 'Failed'
    if (final_copy["Status"].at[i] == 'Failed'):
        result = (r['Designation Level']) - 1    #if 'Failed' then demote 1 level from designation level
        new_designation.append(result)
    else:
        result = (r['Designation Level']) - 0    #if not 'Failed' then do not demote any level from designation level
        new_designation.append(result)

#inserting the new designation column...
final_copy.insert(10, 'Designation', new_designation)     #inserting the Designation column into the final Dataframe at index 10

print("Updated Final DataFrame: ")
final_copy
```

```python
#droping the previous designation Level
final_copy.drop("Designation Level", axis=1 , inplace=True)
```

```python
#updated and renamed Designation Levels

final_copy.rename(columns={'Designation': 'Designation Level'}, inplace=True)

print("Updated Final DataFrame: ")
final_copy
```

```python
#deleting the employee records having designation level > 4

to_delete = 0

for index, row in final_copy.iterrows():
    if row["Designation Level"] > 4:                      #checking if any of the employees have designation level > 4
        to_delete.append(index)                           #if found, then adding the index at the end of the list to_delete      
        final_copy.drop(to_delete, inplace=True)          #removing the index from the Dataframe
print("Updated Final DataFrame: ")
final_copy
```

Task 8 :
Promote designation level by 1 for the employees whose age is more than 29 years using 
IF condition. 

```python
#used iteration of rows and if function to promote designation level by 1 for the employees having age > 29

for i, r in final_copy.iterrows():
    if row["Age"] > 29:
        final_copy.loc[i, 'Designation Level'] += 1       #if age > 29, then promote designation level by 1

    else:
        final_copy.loc[i, 'Designation Level'] += 0       #if age !> 29, then designation level remains same

print("Updated Final DataFrame: ")
final_copy
```

Task 9 :
Add the cost of all projects for each Employee and save it in new dataframe
